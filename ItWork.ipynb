```python
import cv2
from cvzone.HandTrackingModule import HandDetector
import serial.tools.list_ports
import time

# Initialize camera and hand detector
cap = cv2.VideoCapture(0)
detector = HandDetector(maxHands=1, detectionCon=0.7)

# Find Arduino port automatically
def find_arduino_port():
    ports = serial.tools.list_ports.comports()
    for port in ports:
        if "Arduino" in port.description or "USB Serial" in port.description:
            return port.device
    return None

arduino_port = find_arduino_port()
if arduino_port is None:
    print("Arduino not found. Please check your connection.")
    exit()

print(f"Found Arduino at {arduino_port}")

# Initialize serial connection to Arduino
ser = serial.Serial('/dev/cu.usbmodemF412FA7761802', 9600, timeout=1)
time.sleep(2)  # Wait for connection to establish

try:
    while True:
        success, img = cap.read()
        if not success:
            print("Failed to capture image")
            break
            
        hands, img = detector.findHands(img, draw=True, flipType=True)
        
        if hands:  # If there are hands in the camera
            hand1 = hands[0]
            fingers1 = detector.fingersUp(hand1)
            print("Fingers:", fingers1)
            
            # Prepare data string to send to Arduino
            data_string = f"${fingers1[0]},{fingers1[1]},{fingers1[2]},{fingers1[3]},{fingers1[4]}\n"
            ser.write(data_string.encode())
            print("Sent:", data_string.strip())
        else:
            # Send all fingers down if no hand detected
            ser.write(b"$0,0,0,0,0\n")
            print("No hand detected - sending zeros")
        
        cv2.imshow('Hand Tracking', img)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

finally:
    # Cleanup
    cap.release()
    cv2.destroyAllWindows()
    ser.close()
    print("Resources released")
```
